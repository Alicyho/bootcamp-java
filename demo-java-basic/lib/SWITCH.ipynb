```python
！SWITCH！


Example： 
import java.util.Scanner;

public class DemoTzae {


    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        
        System.out.print("請輸入月份 (1-12): ");   // 使用者輸入 11，month = 11 -- "那就會出現 結果 -> November has 30 days."
        int month = scanner.nextInt();
        
        String monthName = "";
        int days = 0;
        
        switch (month) {
            case 1:
                monthName = "January";
                days = 31;
                break;
            case 2:
                monthName = "February";
                days = 28;
                break;
            case 3:
                monthName = "March";
                days = 31;
                break;
            case 4:
                monthName = "April";
                days = 30;
                break;
            case 5:
                monthName = "May";
                days = 31;
                break;
            case 6:
                monthName = "June";
                days = 30;
                break;
            case 7:
                monthName = "July";
                days = 31;
                break;
            case 8:
                monthName = "August";
                days = 31;
                break;
            case 9:
                monthName = "September";
                days = 30;
                break;
            case 10:
                monthName = "October";
                days = 31;
                break;
            case 11:
                monthName = "November";
                days = 30;
                break;
            case 12:
                monthName = "December";
                days = 31;
                break;
            default:
                System.out.println("錯誤：月份必須是 1-12");
                scanner.close();
                return;
        }
        
        System.out.println(monthName + " has " + days + " days");
        scanner.close();
    }

  
}

-----
流程是這樣：
text
第1步：等待使用者輸入數字
       ↓
int month = scanner.nextInt();   // 使用者輸入 5
       ↓
第2步：拿著這個數字 (5) 去 switch 裡面找對應的 case
       ↓
switch (month) {  // month 是 5
    case 1: ...   // 5≠1，跳過
    case 2: ...   // 5≠2，跳過
    case 3: ...   // 5≠3，跳過
    case 4: ...   // 5≠4，跳過
    case 5:       // 5=5！找到了！
        monthName = "May";   // 產生 May
        days = 31;           // 產生 31
        break;
}
       ↓
第3步：輸出結果
System.out.println(monthName + " has " + days + " days");
// 輸出：May has 31 days
一句話總結你的理解：
先等使用者輸入一個數字，
然後 switch 根據這個數字，「產生」對應的月份名稱和天數

-------


第一部分：基礎知識教學
1.1 取得使用者輸入
java
Scanner scanner = new Scanner(System.in);
System.out.print("Input a month number: ");
int month = scanner.nextInt();

1.2 switch-case 語法
java
switch (month) {
    case 1:
        monthOfName = "January";
        number_Of_DaysInMonth = 31;
        break;
    case 2:
        monthOfName = "February";
        number_Of_DaysInMonth = 28;
        break;
    // ... 其他月份
    default:
        System.out.println("Invalid month number.");
}

1.3 閏年公式
java
boolean isLeapYear = (year % 4 == 0 && year % 100 != 0) || (year % 400 == 0);

1.4 只有二月需要調整
java
if (month == 2 && isLeapYear) {
    number_Of_DaysInMonth = 29;
}

1.5 輸出結果
java
System.out.println(monthOfName + " " + year + " has " + number_Of_DaysInMonth + " days");
✍️ 第二部分：練習步驟（請你親自完成）

步驟 1:建立程式基本結構
java
import java.util.Scanner;

public class JavaExercise4 {
    public static void main(String[] args) {
        // 你的程式碼寫在這裡
    }
}
步驟 2:宣告變數
在 main 裡面寫：

java
Scanner scanner = new Scanner(System.in);
int number_Of_DaysInMonth = 0;
String monthOfName = "Unknown";

步驟 3:取得輸入
java
System.out.print("Input a month number: ");
int month = scanner.nextInt();

System.out.print("Input a year: ");
int year = scanner.nextInt();
步驟 4:寫 switch 判斷月份(1 到 12)
請你自己補上 3, 4, 5, 6, 7, 9, 10, 11 的 case。

提示：
31 天:1,3,5,7,8,10,12
30 天:4,6,9,11
28 天:2 (先寫 28)

步驟 5：寫閏年判斷（放在 switch 之後）
java
if (month == 2) {
    boolean isLeapYear = (year % 4 == 0 && year % 100 != 0) || (year % 400 == 0);
    if (isLeapYear) {
        number_Of_DaysInMonth = 29;
    }
}
步驟 6：輸出結果
java
System.out.println(monthOfName + " " + year + " has " + number_Of_DaysInMonth + " days");
scanner.close();
步驟 7：加入 default 處理錯誤輸入
在 switch 最後加上：

java
default:
    System.out.println("Invalid month number.");
    scanner.close();
    return;
🧪 第三部分：測試你的程式
請輸入以下測試資料，確認輸出正確：

輸入月份	輸入年份	預期輸出
2	2016	February 2016 has 29 days
12	2014	December 2014 has 31 days
2	2100	February 2100 has 28 days
2	2000	February 2000 has 29 days
4	2023	April 2023 has 30 days
13	2023	Invalid month number.
⚠️ 第四部分：常見錯誤對照表
錯誤寫法	正確寫法
monthofName = "March"	monthOfName = "March"
number_Of_DaysInMonth = "31"	number_Of_DaysInMonth = 31
else (is not a leap year)	else
開了第二個 Scanner	只用一個 scanner
忘記 break	每個 case 後都要 break
✅ 完成檢查清單
import java.util.Scanner

只有一個 Scanner

switch 有 12 個 case + default

每個 case 都有 break

二月先設 28，再用 if 改為 29

閏年公式正確

輸出格式完全符合題目

scanner.close() 在最後


---------

Switch 主題中的特別注意事項
針對你這個題目，以下是 switch 最容易出錯 的地方：

1. ❌ 忘記 break → 會「掉下去」(fall-through)
java
// 錯誤示範
switch (month) {
    case 1:
        monthOfName = "January";
        number_Of_DaysInMonth = 31;
        // 忘記 break！
    case 2:
        monthOfName = "February";  // 會被執行到！
        number_Of_DaysInMonth = 28;
        break;
}
如果輸入 month = 1，結果會變成 February，因為沒有 break 會繼續往下執行。

2. ❌ 變數名稱大小寫不一致
java
// 錯誤示範
case 3:
    monthofName = "March";  // 小寫 o
    break;

// 之後輸出用的是
System.out.println(monthOfName);  // 大寫 O → 錯誤！
Java 是大小寫敏感的，monthOfName 和 monthofName 是不同變數。

3. ❌ 把數字用引號包起來
java
// 錯誤示範
case 8:
    monthOfName = "August";
    number_Of_DaysInMonth = "31";  // 字串不能放進 int 變數
    break;
number_Of_DaysInMonth 是 int 型態，只能放數字，不能放 "31"。

4. ❌ 忘記處理 default
java
// 不完整示範
switch (month) {
    case 1:
        // ...
        break;
    // 沒有 default
}
如果使用者輸入 13，程式會直接跳過 switch，後面可能產生錯誤輸出。

應該加上：

java
default:
    System.out.println("Invalid month number.");
    scanner.close();
    return;  // 直接結束程式
5. ⚠️ case 可以合併寫（但要注意邏輯）
java
// 合法但不適合這個題目的寫法
case 1:
case 3:
case 5:
    monthOfName = "???";  // 問題：無法區分是哪個月份
    number_Of_DaysInMonth = 31;
    break;
因為你需要不同月份有不同名稱，所以每個 case 必須分開寫。

但天數相同的可以這樣節省程式碼：

java
case 4:
case 6:
case 9:
case 11:
    number_Of_DaysInMonth = 30;
    break;
月份名稱還是要各自處理。

6. ⚠️ switch 可用的資料型態
在 Java 中，switch 可以用在：

int、byte、short、char

String (Java 7+)

enum

這個題目用 int month 是最適合的。

7. 📌 這個題目的 switch 結構建議
java
switch (month) {
    case 1:
        monthOfName = "January";
        number_Of_DaysInMonth = 31;
        break;
    case 2:
        monthOfName = "February";
        number_Of_DaysInMonth = 28;  // 先給平年，後面再用 if 改
        break;
    case 3:
        monthOfName = "March";
        number_Of_DaysInMonth = 31;
        break;
    // ... 4 到 11 依此類推
    case 12:
        monthOfName = "December";
        number_Of_DaysInMonth = 31;
        break;
    default:
        System.out.println("Invalid month number.");
        scanner.close();
        return;
}
快速記憶口訣
case 後要 break
變數大小寫要一致
數字不要加引號
default 記得寫
二月天數先給 28

你把這些注意事項記住，switch 就不會出錯了。有需要我給你一個只有 switch 部分的小練習題嗎？

```
